```python
import numpy as np
import pandas as pd
```

```python
#read training set
train = pd.read_csv('train.csv')
#read testing set
test = pd.read_csv('test.csv')

print('train set:', train.shape, 'test set:', test.shape)
```

```python
#union the two sets for cleaning them quicker
full = train.append(test, ignore_index = True)

print('full set:', full.shape)
```

```python
full.head()
```

```python
full.describe()
```

```python
#check each column of data's type,number of null.
full.info()
```

```python
#fill missing values with mean
#Age
full['Age'] = full['Age'].fillna(full['Age'].mean())
#Fare
full['Fare'] = full['Fare'].fillna(full['Fare'].mean())
```

```python
full['Embarked'].head
```

```python
#fill the two missing value with the value appears most frequently: 'S'
full['Embarked'] = full['Embarked'].fillna('S')
```

```python
full['Cabin'].head
```

```python
#There are many missing values so we fill them with 'U', which means 'unknown'
full['Cabin'] = full['Cabin'].fillna('U')
```

```python
sex_mapDict = {'male':1, 'female':0}
#map: apply function on each data in the series
full['Sex']=full['Sex'].map(sex_mapDict)
full.head()
```

```python
#set a dataframe for storing dummy variables.
embarkedDf = pd.DataFrame()
```

```python
embarkedDf = pd.get_dummies(full['Embarked'], prefix = 'Embarked')
embarkedDf.head()
```

```python
full = pd.concat([full,embarkedDf],axis=1)
full.drop('Embarked',axis=1,inplace=True)
full.head()
```

```python
pclassDf = pd.DataFrame()
pclassDf = pd.get_dummies(full['Pclass'], prefix = 'Pclass')
pclassDf.head()
```

```python
full = pd.concat([full,pclassDf],axis=1)
full.drop('Pclass',axis=1,inplace=True)
full.head()
```

```python
full['Name'].head()
#we want to extract the title (Mr, Miss..) of each person
```

```python
def getTitle(name):
    str1 = name.split(',')[1]
    str2 = str1.split('.')[0]
    str3 = str2.strip()
    return str3
```

```python
titleDf = pd.DataFrame()
titleDf['Title'] = full['Name'].map(getTitle)
titleDf.head()
```

```python
title_mapDict = {
    'Capt': 'Officer',
    'Col': 'Officer',
    'Major': 'Officer',
    'Jonkheer': 'Royalty',
    'Don': 'Royalty',
    'Sir': 'Royalty',
    'Dr': 'Officer',
    'Rev': 'Officer',
    'the Countess': 'Royalty',
    'Dona': 'Royalty',
    'Mme': 'Mrs',
    'Mlle': 'Miss',
    'Ms': 'Mrs',
    'Mr': 'Mr',
    'Mrs': 'Mrs',
    'Miss': 'Miss',
    'Master': 'Master',
    'Lady': 'Royalty'    
}

titleDf['Title'] = titleDf['Title'].map(title_mapDict)

titleDf = pd.get_dummies(titleDf['Title'])
titleDf.head()
```

```python
full = pd.concat([full,titleDf],axis=1)
full.drop('Name',axis=1,inplace=True)
full.head()
```

```python
cabinDf = pd.DataFrame()
full['Cabin'] = full['Cabin'].map(lambda c: c[0])

cabinDf = pd.get_dummies(full['Cabin'], prefix = 'Cabin')
cabinDf.head()
```

```python
full = pd.concat([full,cabinDf],axis=1)
full.drop('Cabin',axis=1,inplace=True)
full.head()
```

```python
familyDf = pd.DataFrame()
familyDf['FamilySize'] = full['Parch']+full['SibSp']+1
#single: size = 1, small = 2 <= size <= 4, large: size >= 5.
familyDf['Family_single'] = familyDf['FamilySize'].map(lambda s : 1 if s == 1 else 0)
familyDf['Family_small'] = familyDf['FamilySize'].map(lambda s : 1 if 2<= s <= 4 else 0)
familyDf['Family_large'] = familyDf['FamilySize'].map(lambda s : 1 if 5 <= s else 0)

familyDf.head()
```

```python
full = pd.concat([full,familyDf],axis=1)
full.head()
```

```python
full.shape
```

```python
corrDf = full.corr()
corrDf
```

```python
corrDf['Survived'].sort_values(ascending = False)
```

```python
full_X = pd.concat( [titleDf,
                     pclassDf,
                     familyDf,
                     full['Fare'],
                     cabinDf,
                     embarkedDf,
                     full['Sex']
                    ] , axis=1 )
full_X.head()
```

```python
sourceRow=891
#features
source_X = full_X.loc[0:sourceRow-1,:]
#label:0 or 1
source_y = full.loc[0:sourceRow-1,'Survived']

pred_X = full_X.loc[sourceRow:,:]
```

```python
print('rows of source set:', source_X.shape[0])

print('rows of predict set:', pred_X.shape[0])
```

```python
'''
从原始数据集（source）中拆分出训练数据集（用于模型训练train），测试数据集（用于模型评估test）
train_test_split是交叉验证中常用的函数，功能是从样本中随机的按比例选取train data和test data
train_data：所要划分的样本特征集
train_target：所要划分的样本结果
test_size：样本占比，如果是整数的话就是样本的数量
'''
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(source_X ,
                                                    source_y,
                                                    train_size=.8)

print ('原始数据集特征：',source_X.shape, 
       '训练数据集特征：',train_X.shape ,
      '测试数据集特征：',test_X.shape)

print ('原始数据集标签：',source_y.shape, 
       '训练数据集标签：',train_y.shape ,
      '测试数据集标签：',test_y.shape)
```

```python
#machine learning algorithm
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

#Random Forests Model
#from sklearn.ensemble import RandomForestClassifier
#model = RandomForestClassifier(n_estimators=100)

#Support Vector Machines
#from sklearn.svm import SVC, LinearSVC
#model = SVC()

#Gradient Boosting Classifier
#from sklearn.ensemble import GradientBoostingClassifier
#model = GradientBoostingClassifier()

#K-nearest neighbors
#from sklearn.neighbors import KNeighborsClassifier
#model = KNeighborsClassifier(n_neighbors = 3)

# Gaussian Naive Bayes
#from sklearn.naive_bayes import GaussianNB
#model = GaussianNB()
```

```python
#model training
model.fit( train_X , train_y)
```

```python
model.score(test_X , test_y)
```

```python
#use this model to predict the test(predict) dataset
pred_Y = model.predict(pred_X)

pred_Y=pred_Y.astype(int)

passenger_id = full.loc[sourceRow:,'PassengerId']

predDf = pd.DataFrame( 
    { 'PassengerId': passenger_id , 
     'Survived': pred_Y } )

predDf.to_csv( 'titanic_pred.csv' , index = False )
```
